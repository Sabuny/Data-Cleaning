* For the **Income** column, lets rename the values of the for easier analysis and interpretation.

```python
#creating the function
def income_group(value):
    if value == "Less than 10000":
        return "<10K"
    elif value == "10001-20000":
        return "10K-20K"
    elif value == "20001-30000":
        return "20K-30K"
    elif value == "30001-40000":
        return "30K-40K"
    elif value == "40001-50000":
        return "40K-50K"
    else:
        return "50K>"

#lets apply  the function
df["Income"] =df["Income"].apply(income_group)

#lets make the column ordered
income_order =["<10K","10K-20K","20K-30K","30K-40K","40K-50K","50K>"]
df["Income"] =pd.Categorical(df["Income"],categories=income_order,ordered=True)

#lets check the new values
print(df["Income"].unique())
```

* Lets create stack bar charts to better understanf the relationship of our features

```python
def plot(df, col, coll):
    stacked = df.groupby([col, coll]).size().unstack()
    ax = stacked.plot(kind="bar", stacked=True, figsize=(10, 6), edgecolor="black")
    for i, bar in enumerate(stacked.iterrows()):
        bottom = 0
        for value, bar_height in enumerate(bar[1]):
            if bar_height > 0:  
                ax.text(
                    i,                    
                    bottom + bar_height / 2, 
                    str(int(bar_height)),    
                    ha='center',             
                    va='center',              
                    fontsize=9,              
                    color="black"         
                )
                bottom += bar_height  # Update the bottom for the next segment
    plt.title(f"{col} vs {coll}",fontweight="bold")
    plt.xlabel(col,fontweight="bold")
    plt.xticks(rotation=45)
    plt.ylabel("Frequency",fontweight="bold")
    plt.grid(axis='y')
    plt.show()
```

## Customer Satisfaction Analysis
Customer satisfaction is a critical factor in business success, influencing customer loyalty, retention, and overall profitability. Understanding customer experiences helps businesses identify strengths, address pain points, and enhance service delivery.

This analysis is based on a quantitative survey conducted by a Nairobi retailer to assess consumer experiences at Store A. The dataset, collected through an online survey in July 2008, contains well-labeled data points, with each row representing an individual respondent.

### Objective
To analyze the survey data and extract actionable insights that can drive improvements in customer satisfaction and overall shopping experience.

* Rename the values for easier readability and interpretation

```python
def count(value):
    if value =="Twice a week":
        return "Twice Weekly"
    elif value =="Thrice a week":
        return "Thrice Weekly"
    elif value =="Twice a month":
        return "Twice Monthly"
    elif value =="I only visitedonce but intend to go back":
        return "Visited Once & Going Back"
    elif value =="Rarely (less than monthly)":
        return "Rarely"
    elif value =="Once a week":
        return "Once Weekly"
    elif value =="Once a month":
        return "Once Monthly"
    elif value =="Thrice a month":
        return "Thrice Monthly"
    elif value =="I only visited once and will not go back":
        return "Visited Once & Not Going Back"
    elif value =="Daily":
        return "Daily"
    else:
        return "Never"

df["Shopping Frequency"] =df["Shopping Frequency"].apply(count)
print(df["Shopping Frequency"].unique())
print("=====" * 20)
print(df["Shopping Frequency"].value_counts())
```

* Rename the values for easier readability and interpretation

```python
def income(value):
    if value == "Less than KES 10,000":
        return "<10K"
    elif value == "KES 10,001 – KES 20,000":
        return "10K-20K"
    elif value == "KES 20,001 – KES 30,000":
        return "20K-30K"
    elif value == "KES 30,001 – KES 40,000":
        return "30K-40K"
    elif value == "KES 40,001 – KES 50,000":
        return "40K-50K"
    elif value == "KES 50,001 – KES 60,000":
        return "50K-60K"
    elif value == "KES 60,001 – KES 70,000":
        return "60K-70K"
    elif value == "KES 70,001 – KES 80,000":
        return "70K-80K"
    elif value == "KES 80,001 – KES 90,000":
        return "80K-90K"
    elif value == "KES 90,001 – KES 100,000":
        return "90K-100K"
    elif value == "Not willing to Answer":
        return np.nan
    else:
        return "100K>"
#Applying the function
df["Monthly Income Group"] = df["Monthly Income"].apply(income)

#Changing the data type to categorical
income_order=["<10K","10K-20K","20K-30K","30K-40K","40K-50K","50K-60K","60K-70K","70K-80K","80K-90K","90K-100K","100K>"]
df["Monthly Income Group"] =pd.Categorical(df["Monthly Income Group"],categories=income_order,ordered=True)
print(df["Monthly Income"].unique())
print("=====" * 20)
print(df["Monthly Income"].value_counts())
```

#### Distribution of Respondents

```python
cols =["Shopping Frequency","Awareness Source","Location Convenience","Product Variety","Favorite Products Availability","Customer Service","Competitive Prices","Promotion Effectiveness","Product Freshness","Fast Checkout","Favorite Store","Other_Stores_1","Score Category"]
for col in cols:
    sns.set_style("whitegrid")
    count =df[col].value_counts(ascending=False)
    order=count.index
    plt.figure(figsize=(10,8))
    ax=sns.countplot(data=df,x=df[col],order=order,edgecolor="black")
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha="center", 
                    va="baseline", 
                    fontsize=10, 
                    color="black", 
                    xytext=(0, 3),
                    textcoords="offset points",
                    fontweight="bold")
    plt.title(f"Distribution of {col}",fontweight="bold")
    plt.ylabel("Frequency",fontweight="bold")
    plt.tight_layout()
    plt.xticks(rotation=45)
    plt.show()
```

#### Insights and Recommendations
##### Shopping Frequency
Most customers shop once a week (66) or once a month (31) while a smaller portion shops daily (17) or three times a week (28).A few customers (13) indicated that they never shop at Store A.
* **Recommendation:** The Store sales manager should encourage more frequent shopping by introducing weekly promotions, discounts and loyalty rewards for regular customers.Also the stakeholders can investigate reasons behind rare or never shopping customers to identify and address barriers.
##### Awareness Source
The primary awareness channel is on the road (143), followed by billboard ads (32) and online (27).The referrals from family/friends (15) are lower than expected.
* **Recommendation:** The marketing could strengthen digital marketing efforts to increase awareness through social media, email marketing and influencer collaborations.Moreover,introduce referral programs where existing customers get discounts for bringing in new customers.
##### Location Convenience
Majority of customers are very satisfied (154) or satisfied (41) with Store A’s location.Only a small number (12) are neutral, while 12 customers both the unsatisfied and very unsatisfied,find the location inconvenient.
* **Recommendation:** Leverage the store’s good location by highlighting it in marketing campaigns and gathering feedback from dissatisfied customers to understand specific accessibility issues (e.g., parking, traffic, store visibility).
##### Product Variety & Availability
Product variety is well-rated, with 110 customers very satisfied and 53 satisfied. Availability of favorite products has similar ratings (106 very satisfied, 56 satisfied).
* **Recommendation:** Maintain a diverse product range and ensure restocking efficiency.Use customer purchase data to forecast demand and prevent stockouts of popular items.
##### Customer Service
Customer service has high satisfaction levels (139 very satisfied, 44 satisfied).A total of 16 customers were dissatisfied with the customer service.
* **Recommendation:** The manager should encourage continuous employee trainee on excellent customer service to maintain high satisfaction and investigate complaints from dissatisfied customers to understand specific service issues.
##### Competitive Prices
From our distribution plot,91 customers are very satisfied, 64 satisfied with pricing, but 28 are dissatisfied.
* **Recommendation:** Maintaining competitive pricing and offering price-match guarantees where feasible can attract more customers.Also introducing exclusive discounts for loyal customers can improve satisfaction with pricing.
##### Promotion Effectiveness
102 customers were very satisfied, 61 satisfied, indicating promotions are mostly effective while 17 customers were dissatisfied.
* **Recommendation:** Enhance visibility of promotions through in-store displays, SMS alerts, and social media campaigns while experimenting with different types of promotions (e.g., buy-one-get-one-free, flash sales, or limited-time discounts).
##### Product Freshness
125 customers were very satisfied, 51 satisfied customers indicate freshness is a strong point leaving 18 customers dissatisfied.
* **Recommendation:** Maintain high freshness standards by improving supply chain efficiency and inventory management.Also use customer feedback to identify specific freshness concerns (e.g., fruits, vegetables, dairy).
##### Fast Checkout
Checkout experience is mostly positive with 132 customers very satisfied and 39 satisfied while 23 customers were dissatisfied.
* **Recommendation:** Training cashiers on speed and efficiency to minimize waiting times.
##### Competitor Stores & Market Positioning
Carrefour – TRM (91) and Quickmart – Mirema (29) are the main competitors although Store A is the favorite for 167 respondents, followed by Carrefour – TRM with 25 customers.
* **Recommendation:** Analyze competitor strategies and price comparisons and offer unique value propositions to retain market dominance or adjust promotion prices.
### Demographics (Gender & Age Group)
A majority of customers are females and highest age groups are customers age between 31-35 and 26-30.
* **Recommendation:** The marketin team should tailor their efforts towards women and millennials (26-35 age group) and also introduce gender-specific promotions or personalized product recommendations based on shopping habits.
##### Recommendation Score
174 customers gave a high score, 30 gave an average score, and 15 gave a low score.
* **Recommendation:** Request satisfied customers to leave online reviews to attract more shoppers.
##### Monthly Income Distribution
Majority of customers earn 10K-20K (76), 30K-40K (29) and 20K-30K (22) whereas few customers earn above 100K (17).
* **Recommendation:** Offer affordable pricing and budget-friendly promotions targeting middle-income earners Introduce premium product lines for higher-income groups.

#### Final Takeaways
* ✅ Leverage strengths: Strong location, good customer service, high product freshness and fast checkout.
* ✅ Address weaknesses: Improve pricing perception, enhance promotions and optimize customer service.
* ✅ Enhance engagement: Use referral programs, personalized offers and loyalty incentives to boost customer retention.

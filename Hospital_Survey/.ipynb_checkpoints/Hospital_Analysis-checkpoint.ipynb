```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
df =pd.read_csv("Hospital_Data.csv")
display(df.head())
print(f"\nShape of the dataset: {df.shape}")
print(f"\nDataset data types: {df.dtypes}")
print(f"\nMissing values present: {df.isna().sum()}")
```

#### Data cleaning/preprocessing

```python
#Renaming the columns
df.columns=[
    "Id",
    "Admission date",
    "First intial name",
    "Last name",
    "Gender",
    "Age",
    "Race",
    "Department referral",
    "Admission flag",
    "Satisfaction score",
    "Waiting time",
    "Patient CM"
]

#Change the Admission date data type
df["Admission date"]=pd.to_datetime(df["Admission date"],format="%d-%m-%Y %H:%M")

#Imputing missing values
df["Department referral"] =df["Department referral"].fillna(df["Department referral"].mode()[0])
df["Satisfaction score"]=df["Satisfaction score"].fillna(df["Satisfaction score"].mean())

display(df.head())
print(f"\nShape of the dataset: {df.shape}")
print(f"\nDataset data types: {df.dtypes}")
print(f"\nMissing values present: {df.isna().sum()}")
```

```python
race=df["Race"].value_counts()
ax=race.plot(kind="barh",edgecolor="black")
ax.set_xlabel("Count",fontweight="bold")
ax.set_ylabel("Race",fontweight="bold")
ax.invert_yaxis()
ax.grid(axis="x",linestyle="--")
ax.set_title("Distribution of Gender",fontweight="bold")
plt.show();
```

#### Department & Admission Trends

```python
depart=df["Department referral"].value_counts()
ax=depart.plot(kind="barh",edgecolor="black")
ax.set_ylabel("Department referral",fontweight="bold")
ax.set_xlabel("Count",fontweight="bold")
ax.set_title("Department referral distribution",fontweight="bold")
ax.invert_yaxis()
```

##### Admissions per department

```python
dept=df.groupby("Department referral")["Admission flag"].sum().sort_values(ascending=False)
ax=dept.plot(kind="barh",edgecolor="black")
ax.set_ylabel("Department referral",fontweight="bold")
ax.set_xlabel("Count",fontweight="bold")
ax.set_title("Distribution of Admissions per Deparment",fontweight="bold")
ax.invert_yaxis()
```

##### Average age per department

```python
age=df.groupby("Department referral")["Age"].mean().sort_values(ascending=False)
ax=age.plot(kind="barh",edgecolor="black")
ax.set_ylabel("Department referral",fontweight="bold")
ax.set_xlabel("Average Age",fontweight="bold")
ax.set_title("Distribution of Average Age per Department",fontweight="bold")
ax.invert_yaxis()
```

#### Patient Satisfaction Analysis


```python
sns.histplot(df["Satisfaction score"], bins=10,kde=True,edgecolor="black")
plt.xlabel("Satisfaction Score",fontweight="bold")
plt.ylabel("Count",fontweight="bold")
plt.title("Satisfaction Score Distribution",fontweight="bold")
plt.show()
```

##### Satisfaction by department:

```python
sat=df.groupby("Department referral")["Satisfaction score"].mean().sort_values(ascending=False)
ax=sat.plot(kind="barh")
ax.set_title("Distribution of Average Satisfaction by Department referral",fontweight="bold")
ax.set_xlabel("Average Satisfaction Score",fontweight="bold")
ax.set_ylabel("Department Referral",fontweight="bold")
ax.invert_yaxis()
```

##### Satisfaction by waiting time 

```python
correlation, p_value = stats.pearsonr(df["Waiting time"], df["Satisfaction score"])
print(f"Correlation: {correlation}, P-value: {p_value}")
```

A strong negative correlation means longer waiting times reduce satisfaction.
